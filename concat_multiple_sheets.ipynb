```python
import pandas as pd
```

```python
xls = pd.ExcelFile("/content/pe_01.xlsx")
pd.read_excel(xls, 'Sheet2')
```

```python
df = pd.DataFrame()
for i in xls.sheet_names:
  sheet = pd.read_excel(xls, i, header=None)
  sheet.dropna(axis=0, how='all', inplace=True)
  sheet.dropna(axis=1, how='all', inplace=True)
  sheet.columns = sheet.iloc[0]
  sheet = sheet.iloc[1:]
  df = pd.concat([df, sheet], ignore_index=True)

df
```
